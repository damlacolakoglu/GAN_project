```python
from PIL import Image
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def resize_image(image_path, target_size):
    image = Image.open(image_path)
    resized_image = image.resize(target_size)
    return resized_image
```

```python
def normalize_image(image):
    return np.array(image) / 255.0
```

```python
def prepare_dataset(dataset_dir, target_size=(128, 128)):
    images = []
    for class_name in os.listdir(dataset_dir):
        class_dir = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_dir):  # Sadece dizinleri işlemek için kontrol ekledik
            for filename in os.listdir(class_dir):
                if filename.endswith(".jpg"):
                    image_path = os.path.join(class_dir, filename)
                    resized_image = resize_image(image_path, target_size)
                    normalized_image = normalize_image(resized_image)
                    images.append(normalized_image)
    
    return np.array(images)
```

```python
# Verilerin olduğu ana klasörün yolu
dataset_dir = "dataset3"
# Veri boyutunu belirleme
target_size = (128, 128)
```

```python
dataset = prepare_dataset(dataset_dir, target_size)
np.save('dataset1.npy', dataset)
```
